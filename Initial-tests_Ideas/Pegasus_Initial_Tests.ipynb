# 5.Evaluation

```python
from transformers import AutoTokenizer, AutoModelWithLMHead, AutoModelForSeq2SeqLM
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu
# Define los modelos que quieres probar
model_names = ["google/pegasus-xsum", "facebook/bart-large-cnn", "t5-base", "microsoft/prophetnet-large-uncased"]

# Define el texto que quieres resumir
#https://en.wikipedia.org/wiki/Bruno_Kova%C4%8Di%C4%87
text = """His career began in the 1980s when he formed the band called D-Day. Afterwards he worked with all the important musicians and singers in and around Croatia as a musician, composer or an arranger. By late 80'a he entered Parni valjak as a guitarist and saxophone player, then performed with Gibonni, Hard Time and Fanny Hill, and he recorded his album "Just Like All Other People" in 2001. He wrote songs for Oliver Dragojević, Ivan Mikulić, Tony Cetinski, Danijela, Vanna, Jacques Houdek, and for his wife Ivana Plechinger with whom he wrote hit songs such as To nismo mi and Kao rijeka which scored second on Dora in 2000. Ivana and Bruno have two sons."""
reference_summary = """In the 1980s, he founded the band D-Day, becoming a prominent musician, composer, and arranger in Croatia. Joining Parni valjak, he played guitar and saxophone, collaborating with artists like Gibonni. His 2001 album "Just Like All Other People" showcased his talents. Known for hit songs, he wrote for artists like Oliver Dragojević and his wife Ivana Plechinger. Bruno and Ivana have two sons."""
# Inicializa el evaluador de ROUGE
scorer = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)

print(f"Texto Original: {text}\n")
print(f"Reference Summary: {reference_summary}")
print("_________________________________________________________________\n\n")
# Para cada modelo
for model_name in model_names:
    # Carga el tokenizador y el modelo
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    if (model_name=="google/pegasus-xsum" or model_name=="microsoft/prophetnet-large-uncased"):
      model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
    else:
      model = AutoModelWithLMHead.from_pretrained(model_name)

    # Tokeniza el texto
    inputs = tokenizer.encode("summarize: " + text, return_tensors='pt', max_length=512, truncation=True)

    # Genera el resumen
    summary_ids = model.generate(inputs, max_length=150, min_length=40, length_penalty=5., num_beams=2)
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    # Calcula las puntuaciones de ROUGE
    scores = scorer.score(summary,reference_summary)

    # Calcula la puntuación BLEU
    reference = reference_summary.split()
    candidate = summary.split()
    bleu_score = sentence_bleu([reference], candidate)

    # Imprime las puntuaciones
    print("_________________________________________________________________")
    print(f"Modelo: {model_name}")
    print(f"Resumen: {summary}")

    rouge1_score = scores['rouge1']
    rouge2_score = scores['rouge2']
    rougeL_score = scores['rougeL']

    print(f"Puntuación ROUGE-1: {rouge1_score}")
    print(f"Puntuación ROUGE-2: {rouge2_score}")
    print(f"Puntuación ROUGE-L: {rougeL_score}")
    print(f"Puntuación BLEU: {bleu_score}")
    print("_________________________________________________________________")
    print("\n")
```

# 6. Test Long Text


```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer
from concurrent.futures import ThreadPoolExecutor
import re
import torch

# Cargar el tokenizador y el modelo
model_name = "tuner007/pegasus_summarizer"
tokenizer = PegasusTokenizer.from_pretrained(model_name)
model = PegasusForConditionalGeneration.from_pretrained(model_name)

# Mover el modelo a la GPU si está disponible
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)

def divide_texto(texto, longitud_max=512):
    parrafos = re.split('\n|\.', texto)
    trozos = []
    trozo_actual = ''
    for i, parrafo in enumerate(parrafos):
        if len(tokenizer.encode(trozo_actual + '\n' + parrafo)) <= longitud_max:
            trozo_actual += '\n' + parrafo
        else:
            trozos.append((i, trozo_actual))
            trozo_actual = parrafo
    trozos.append((i, trozo_actual))
    return trozos

def resumir_texto(trozo):
    id, texto = trozo
    # Codificar el texto y generar el resumen
    inputs = tokenizer.encode(texto, return_tensors='pt', truncation=True, padding='longest').to(device)
    summary_ids = model.generate(inputs, max_length=150, min_length=20, length_penalty=2, num_beams=4, early_stopping=True)
    resumen = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return id, resumen

def resumir_texto_final(trozo):
    id, texto = trozo
    # Codificar el texto y generar el resumen
    inputs = tokenizer.encode(texto, return_tensors='pt', truncation=True, padding='longest').to(device)
    summary_ids = model.generate(inputs, max_length=400, min_length=60, length_penalty=2.0, num_beams=16, early_stopping=True, no_repeat_ngram_size=4)
    resumen = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return id, resumen

def resumir_texto_paralelo(texto, longitud_max):
    # Dividir el texto en trozos
    trozos = divide_texto(texto, longitud_max)

    resumenes = {}
    for i, trozo in enumerate(trozos):
        id, resumen = resumir_texto(trozo)
        resumenes[id] = resumen
        porcentaje_completado = ((i + 1) / len(trozos)) * 100
        print(f'Proceso completado: {i + 1}. Porcentaje completado: {porcentaje_completado}%')
        print(f'Resumen: {resumen}')

    # Ordenar los resúmenes por id y juntarlos
    texto_resumen = ' '.join(resumen for id, resumen in sorted(resumenes.items()))

    # Resumir el texto resumen
    print(texto_resumen)
    resumen_final = resumir_texto_final((None, texto_resumen))[1]
    return resumen_final


# Texto de prueba
texto = texto_traducido

# Longitud máxima de cada trozo
longitud_max = 450

# Generar el resumen
resumen_final = resumir_texto_paralelo(texto, longitud_max)

# Imprimir el resumen final
print("\n",resumen_final)
```

```python
from transformers import PegasusForConditionalGeneration, PegasusTokenizer

# Cargar el tokenizador y el modelo para parafrasear
tokenizer_par = PegasusTokenizer.from_pretrained("tuner007/pegasus_paraphrase")
model_par = PegasusForConditionalGeneration.from_pretrained("tuner007/pegasus_paraphrase")

def parafrasear(texto):
    # Codificar el texto y generar la parafrase
    inputs = tokenizer_par.encode(texto, return_tensors='pt', truncation=True, padding='longest').to(device)
    paraphrase_ids = model_par.generate(inputs, max_length=150, min_length=20, length_penalty=2, num_beams=4, early_stopping=True)
    parafraseo = tokenizer_par.decode(paraphrase_ids[0], skip_special_tokens=True)
    return parafraseo

def resumir_y_parafrasear(texto, longitud_max):
    # Dividir el texto en trozos
    trozos = divide_texto(texto, longitud_max)

    resumenes = {}
    for i, trozo in enumerate(trozos):
        id, resumen = resumir_texto(trozo)
        resumenes[id] = resumen
        porcentaje_completado = ((i + 1) / len(trozos)) * 100
        print(f'Proceso completado: {i + 1}. Porcentaje completado: {porcentaje_completado}%')
        print(f'Resumen: {resumen}')

    # Ordenar los resúmenes por id y juntarlos
    texto_resumen = ' '.join(resumen for id, resumen in sorted(resumenes.items()))

    # Resumir el texto resumen
    print(texto_resumen)
    resumen_final = resumir_texto_final((None, texto_resumen))[1]

    # Parafrasear el resumen final
    parafraseo_final = parafrasear(resumen_final)

    return parafraseo_final

# Generar el resumen parafraseado
resumen_parafraseado_final = resumir_y_parafrasear(texto, longitud_max)

# Imprimir el resumen parafraseado final
print("\n", resumen_parafraseado_final)
```
